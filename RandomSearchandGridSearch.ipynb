```python
from numpy.random import rand

# función objetivo
def objective(x):
    return x**2.0

# definir rango para la entrada
r_min, r_max = -5.0, 5.0

# generar una muestra aleatoria del dominio (100 valores)
sample = r_min + rand(100) * (r_max - r_min)

# evaluar la muestra
sample_eval = objective(sample)

# localizar la mejor solución (mínimo)
best_ix = 0
for i in range(len(sample)):
    if sample_eval[i] < sample_eval[best_ix]:
        best_ix = i

# resumen de la mejor solución
print('Best: f(%.5f) = %.5f' % (sample[best_ix], sample_eval[best_ix]))
```

```python
from numpy import arange
from numpy.random import rand
from matplotlib import pyplot

# función objetivo
def objective(x):
    return x**2.0

# definir rango para la entrada
r_min, r_max = -5.0, 5.0

# generar una muestra aleatoria del dominio (100 valores)
sample = r_min + rand(100) * (r_max - r_min)

# evaluar la muestra
sample_eval = objective(sample)

# localizar la mejor solución (mínimo)
best_ix = 0
for i in range(len(sample)):
    if sample_eval[i] < sample_eval[best_ix]:
        best_ix = i

# imprimir la mejor solución
print('Best: f(%.5f) = %.5f' % (sample[best_ix], sample_eval[best_ix]))

# rango de entrada muestreado uniformemente con incrementos de 0.1
inputs = arange(r_min, r_max, 0.1)

# calcular resultados
results = objective(inputs)

# gráfico línea de input vs resultado
pyplot.plot(inputs, results)

# graficar las muestras
pyplot.scatter(sample, sample_eval)

# línea vertical en la mejor solución encontrada
pyplot.axvline(x=sample[best_ix], ls='--', color='red')

# mostrar gráfico
pyplot.show()
```

```python
from numpy import arange
from numpy.random import rand

# función objetivo
def objective(x, y):
    return x**2.0 + y**2.0

# definir rango para la entrada
r_min, r_max = -5.0, 5.0

# generar una malla (grid) de muestras en el dominio
sample = list()
step = 0.1
for x in arange(r_min, r_max + step, step):
    for y in arange(r_min, r_max + step, step):
        sample.append([x, y])

# evaluar cada muestra
sample_eval = [objective(x, y) for x, y in sample]

# localizar la mejor solución (mínimo)
best_ix = 0
for i in range(len(sample)):
    if sample_eval[i] < sample_eval[best_ix]:
        best_ix = i

# mostrar la mejor solución encontrada
print('Best: f(%.5f, %.5f) = %.5f' % (sample[best_ix][0], sample[best_ix][1], sample_eval[best_ix]))
```

```python
from numpy import arange
from numpy import meshgrid
from numpy.random import rand
from matplotlib import pyplot

# función objetivo
def objective(x, y):
    return x**2.0 + y**2.0

# definir rango para la entrada
r_min, r_max = -5.0, 5.0

# generar una malla (grid) de muestras en el dominio
sample = list()
step = 0.5
for x in arange(r_min, r_max + step, step):
    for y in arange(r_min, r_max + step, step):
        sample.append([x, y])

# evaluar cada muestra
sample_eval = [objective(x, y) for x, y in sample]

# localizar la mejor solución (mínimo)
best_ix = 0
for i in range(len(sample)):
    if sample_eval[i] < sample_eval[best_ix]:
        best_ix = i

# mostrar la mejor solución encontrada
print('Best: f(%.5f, %.5f) = %.5f' % (sample[best_ix][0], sample[best_ix][1], sample_eval[best_ix]))

# rango de entrada muestreado uniformemente con incrementos de 0.1
xaxis = arange(r_min, r_max, 0.1)
yaxis = arange(r_min, r_max, 0.1)

# crear una malla a partir de los ejes
x, y = meshgrid(xaxis, yaxis)

# calcular resultados para la malla
results = objective(x, y)

# crear un gráfico de contornos rellenos
pyplot.contourf(x, y, results, levels=50, cmap='jet')

# graficar las muestras como puntos negros
pyplot.plot([p[0] for p in sample], [p[1] for p in sample], '.', color='black')

# marcar la mejor solución con una estrella blanca
pyplot.plot(sample[best_ix][0], sample[best_ix][1], '*', color='white')

# mostrar gráfico
pyplot.show()
```
