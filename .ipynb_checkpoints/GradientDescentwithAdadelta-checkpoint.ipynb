```python
from numpy import asarray, arange, meshgrid
from matplotlib import pyplot as plt

# funcin objetivo
def objective(x, y):
    return x**2.0 + y**2.0

# definir rango de entrada
bounds = asarray([[-1.0, 1.0], [-1.0, 1.0]])

# generar valores uniformes en el rango con paso 0.1
xaxis = arange(bounds[0,0], bounds[0,1] + 0.1, 0.1)
yaxis = arange(bounds[1,0], bounds[1,1] + 0.1, 0.1)

# crear malla 2D
x, y = meshgrid(xaxis, yaxis)

# calcular valores de la funcin objetivo
results = objective(x, y)

# crear grfico de contorno relleno con 50 niveles y colormap 'jet'
plt.contourf(x, y, results, levels=50, cmap='jet')

# mostrar grfico
plt.show()
```

```python
from math import sqrt
from numpy import asarray
from numpy.random import rand, seed

# funcin objetivo
def objective(x, y):
    return x**2.0 + y**2.0

# derivada de la funcin objetivo
def derivative(x, y):
    return asarray([2.0 * x, 2.0 * y])

# algoritmo de descenso por gradiente con Adadelta
def adadelta(objective, derivative, bounds, n_iter, rho, ep=1e-3):
    # generar un punto inicial aleatorio dentro de los bounds
    solution = bounds[:, 0] + rand(len(bounds)) * (bounds[:, 1] - bounds[:, 0])
    # lista de promedios cuadrados de gradientes para cada variable
    sq_grad_avg = [0.0 for _ in range(bounds.shape[0])]
    # lista de promedios cuadrados de actualizaciones para cada variable
    sq_para_avg = [0.0 for _ in range(bounds.shape[0])]

    # ciclo de iteraciones
    for it in range(n_iter):
        # calcular gradiente en el punto actual
        gradient = derivative(solution[0], solution[1])

        # actualizar promedio mvil de cuadrados de gradientes
        for i in range(gradient.shape[0]):
            sg = gradient[i]**2.0
            sq_grad_avg[i] = (sq_grad_avg[i] * rho) + (sg * (1.0 - rho))

        # construir la nueva solucin variable por variable
        new_solution = list()
        for i in range(solution.shape[0]):
            # calcular tamao del paso para la variable i
            alpha = (ep + sqrt(sq_para_avg[i])) / (ep + sqrt(sq_grad_avg[i]))
            # calcular el cambio en la variable
            change = alpha * gradient[i]
            # actualizar promedio mvil de cuadrados de cambios
            sq_para_avg[i] = (sq_para_avg[i] * rho) + (change**2.0 * (1.0 - rho))
            # actualizar el valor de la variable
            value = solution[i] - change
            new_solution.append(value)

        # actualizar la solucin actual con la nueva
        solution = asarray(new_solution)
        # evaluar la funcin objetivo en la nueva solucin
        solution_eval = objective(solution[0], solution[1])

        # imprimir progreso
        print(f'>{it} f({solution}) = {solution_eval:.5f}')

    return [solution, solution_eval]

# fijar semilla para reproducibilidad
seed(1)

# definir lmites del espacio de bsqueda
bounds = asarray([[-1.0, 1.0], [-1.0, 1.0]])

# definir nmero de iteraciones
n_iter = 120

# parmetro momentum para Adadelta
rho = 0.99

# ejecutar descenso por gradiente con Adadelta
best, score = adadelta(objective, derivative, bounds, n_iter, rho)

print('Done!')
print(f'f({best}) = {score}')
```

```python
from math import sqrt
from numpy import asarray, arange, meshgrid
from numpy.random import rand, seed
from matplotlib import pyplot as plt

# funcin objetivo
def objective(x, y):
    return x**2.0 + y**2.0

# derivada de la funcin objetivo
def derivative(x, y):
    return asarray([2.0 * x, 2.0 * y])

# descenso por gradiente con Adadelta que guarda todas las soluciones
def adadelta(objective, derivative, bounds, n_iter, rho, ep=1e-3):
    solutions = list()
    # punto inicial aleatorio dentro de los lmites
    solution = bounds[:, 0] + rand(len(bounds)) * (bounds[:, 1] - bounds[:, 0])
    # promedios mviles de gradientes y parmetros al cuadrado
    sq_grad_avg = [0.0 for _ in range(bounds.shape[0])]
    sq_para_avg = [0.0 for _ in range(bounds.shape[0])]

    for it in range(n_iter):
        gradient = derivative(solution[0], solution[1])
        # actualizar promedio mvil del cuadrado de gradientes
        for i in range(len(gradient)):
            sg = gradient[i] ** 2.0
            sq_grad_avg[i] = sq_grad_avg[i] * rho + sg * (1.0 - rho)
        # calcular nueva solucin
        new_solution = []
        for i in range(len(solution)):
            alpha = (ep + sqrt(sq_para_avg[i])) / (ep + sqrt(sq_grad_avg[i]))
            change = alpha * gradient[i]
            sq_para_avg[i] = sq_para_avg[i] * rho + (change ** 2.0) * (1.0 - rho)
            value = solution[i] - change
            new_solution.append(value)
        solution = asarray(new_solution)
        solutions.append(solution)
        solution_eval = objective(solution[0], solution[1])
        print(f'>{it} f({solution}) = {solution_eval:.5f}')
    return solutions

# semilla para reproducibilidad
seed(1)

# lmites del espacio de bsqueda
bounds = asarray([[-1.0, 1.0], [-1.0, 1.0]])

# nmero de iteraciones
n_iter = 120

# parmetro rho para Adadelta
rho = 0.99

# ejecutar descenso por gradiente con Adadelta
solutions = adadelta(objective, derivative, bounds, n_iter, rho)

# preparar datos para grfico de contorno
xaxis = arange(bounds[0,0], bounds[0,1] + 0.1, 0.1)
yaxis = arange(bounds[1,0], bounds[1,1] + 0.1, 0.1)
x, y = meshgrid(xaxis, yaxis)
results = objective(x, y)

# graficar contorno relleno
plt.contourf(x, y, results, levels=50, cmap='jet')

# graficar trayectoria de soluciones (puntos y lneas blancas)
solutions = asarray(solutions)
plt.plot(solutions[:, 0], solutions[:, 1], '.-', color='w')

plt.title('Descenso por gradiente con Adadelta sobre f(x,y)=x+y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
