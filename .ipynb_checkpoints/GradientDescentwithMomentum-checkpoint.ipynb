```python
from numpy import asarray
from numpy.random import rand
from numpy.random import seed

# Función objetivo
def objective(x):
    return x**2.0

# Derivada de la función objetivo
def derivative(x):
    return x * 2.0

# Algoritmo de descenso de gradiente
def gradient_descent(objective, derivative, bounds, n_iter, step_size):
    # Generar un punto inicial aleatorio dentro de los límites
    solution = bounds[:, 0] + rand(len(bounds)) * (bounds[:, 1] - bounds[:, 0])
    
    # Ejecutar las iteraciones del descenso de gradiente
    for i in range(n_iter):
        # Calcular el gradiente en la posición actual
        gradient = derivative(solution)
        
        # Actualizar la posición dando un paso en dirección opuesta al gradiente
        solution = solution - step_size * gradient
        
        # Evaluar la función objetivo en la nueva posición
        solution_eval = objective(solution)
        
        # Imprimir progreso
        print('>%d f(%s) = %.5f' % (i, solution, solution_eval))
    
    return [solution, solution_eval]

# Fijar la semilla del generador aleatorio para reproducibilidad
seed(4)

# Definir rango para la variable de entrada (aquí solo 1 dimensión: de -1 a 1)
bounds = asarray([[-1.0, 1.0]])

# Definir el número total de iteraciones
n_iter = 30

# Definir el tamaño del paso
step_size = 0.1

# Ejecutar la búsqueda con descenso de gradiente
best, score = gradient_descent(objective, derivative, bounds, n_iter, step_size)

print('Done!')
print('f(%s) = %f' % (best, score))
```

```python
from numpy import asarray, arange
from numpy.random import rand, seed
from matplotlib import pyplot as plt

# Función objetivo
def objective(x):
    return x**2.0

# Derivada de la función objetivo
def derivative(x):
    return x * 2.0

# Algoritmo de descenso de gradiente
def gradient_descent(objective, derivative, bounds, n_iter, step_size):
    # Guardar todas las soluciones y sus valores
    solutions, scores = list(), list()
    # Generar un punto inicial aleatorio dentro de los límites
    solution = bounds[:, 0] + rand(len(bounds)) * (bounds[:, 1] - bounds[:, 0])
    
    # Ejecutar las iteraciones
    for i in range(n_iter):
        # Calcular gradiente en la posición actual
        gradient = derivative(solution)
        # Actualizar la posición
        solution = solution - step_size * gradient
        # Evaluar la función objetivo
        solution_eval = objective(solution)
        # Guardar solución y evaluación
        solutions.append(solution.copy())  # .copy() para evitar referencias
        scores.append(solution_eval)
        # Mostrar progreso
        print('>%d f(%s) = %.5f' % (i, solution, solution_eval))
    
    return [solutions, scores]

# Semilla para reproducibilidad
seed(4)

# Rango para la variable de entrada (1 dimensión, de -1 a 1)
bounds = asarray([[-1.0, 1.0]])

# Número total de iteraciones
n_iter = 30

# Tamaño del paso
step_size = 0.1

# Ejecutar descenso de gradiente
solutions, scores = gradient_descent(objective, derivative, bounds, n_iter, step_size)

# Preparar rango de entrada para graficar la función
inputs = arange(bounds[0, 0], bounds[0, 1] + 0.1, 0.1)
results = objective(inputs)

# Graficar la función objetivo
plt.plot(inputs, results, label='Función objetivo')

# Graficar el recorrido del descenso de gradiente
plt.plot(solutions, scores, '.-', color='red', label='Descenso de gradiente')

plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Visualización de Descenso de Gradiente')
plt.show()
```

```python
from numpy import asarray
from numpy.random import rand
from numpy.random import seed

# Función objetivo
def objective(x):
    return x**2.0

# Derivada de la función objetivo
def derivative(x):
    return x * 2.0

# Descenso de gradiente con momentum
def gradient_descent(objective, derivative, bounds, n_iter, step_size, momentum):
    # Generar punto inicial aleatorio
    solution = bounds[:, 0] + rand(len(bounds)) * (bounds[:, 1] - bounds[:, 0])
    # Mantener el cambio anterior (inicialmente 0)
    change = 0.0
    
    # Iterar n_iter veces
    for i in range(n_iter):
        # Calcular gradiente
        gradient = derivative(solution)
        # Calcular nuevo cambio con momentum
        new_change = step_size * gradient + momentum * change
        # Actualizar posición
        solution = solution - new_change
        # Guardar cambio para la siguiente iteración
        change = new_change
        # Evaluar función objetivo
        solution_eval = objective(solution)
        # Mostrar progreso
        print('>%d f(%s) = %.5f' % (i, solution, solution_eval))
    
    return [solution, solution_eval]

# Semilla para reproducibilidad
seed(4)

# Rango para la variable
bounds = asarray([[-1.0, 1.0]])

# Número de iteraciones
n_iter = 30

# Tamaño del paso
step_size = 0.1

# Parámetro de momentum
momentum = 0.3

# Ejecutar descenso con momentum
best, score = gradient_descent(objective, derivative, bounds, n_iter, step_size, momentum)

print('Done!')
print('f(%s) = %f' % (best, score))
```
