```python
from numpy import arange
from matplotlib import pyplot as plt

# Función objetivo 1D
def objective(x):
    return x**2.0

# Rango de entrada
r_min, r_max = -5.0, 5.0

# Muestreo uniforme con incrementos de 0.1
inputs = arange(r_min, r_max, 0.1)

# Evaluar función objetivo
results = objective(inputs)

# Mostrar algunos valores
print('Inputs:', inputs[:5])
print('Results:', results[:5])

# Mapeo inputs a outputs
for i in range(5):
    print(f'f({inputs[i]:.3f}) = {results[i]:.3f}')
```

```python
# Línea de la función
plt.plot(inputs, results, label='Función objetivo')

# Punto óptimo conocido (mínimo en x=0)
optima_x = 0.0
optima_y = objective(optima_x)
plt.plot([optima_x], [optima_y], 's', color='r', label='Óptimo')

# Muestra la gráfica
plt.legend()
plt.title('Función Objetivo 1D')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```python
plt.scatter(inputs, results, color='black', label='Muestras')

plt.title('Scatter plot función objetivo 1D')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

```python
from numpy.random import seed, rand

# Semilla para reproducibilidad
seed(1)

# Generar 10 muestras aleatorias uniformes en el rango
sample = r_min + rand(10) * (r_max - r_min)

# Evaluar función objetivo en las muestras
sample_eval = objective(sample)

# Graficar línea función y muestras
plt.plot(inputs, results, label='Función objetivo')
plt.plot(sample, sample_eval, 'o', color='black', label='Muestras aleatorias')

# Óptimo marcado con línea roja discontinua
plt.axvline(x=optima_x, ls='--', color='red', label='Óptimo')

plt.title('Función objetivo 1D con muestras de optimización')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

```python
from numpy import arange, meshgrid

# Función objetivo 2D
def objective_2d(x, y):
    return x**2 + y**2

# Rango de entrada
r_min, r_max = -5.0, 5.0

# Crear rangos para x e y
xaxis = arange(r_min, r_max, 0.1)
yaxis = arange(r_min, r_max, 0.1)

# Crear meshgrid
x, y = meshgrid(xaxis, yaxis)

print('X meshgrid sample:')
print(x[:5, :5])

print('Y meshgrid sample:')
print(y[:5, :5])

# Evaluar función en todos los puntos del meshgrid
results_2d = objective_2d(x, y)

print('Resultados sample:')
print(results_2d[:5, :5])

# Mostrar algunos mapeos de entrada a salida
for i in range(5):
    print(f'f({x[i,0]:.3f}, {y[i,0]:.3f}) = {results_2d[i,0]:.3f}')
```

```python
plt.contourf(x, y, results_2d, levels=50, cmap='viridis')
plt.colorbar(label='f(x,y)')
plt.title('Mapa de calor función objetivo 2D')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x, y, results_2d, cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x,y)')
ax.set_title('Superficie 3D función objetivo 2D')

plt.show()
```

```python
from numpy import arange
from numpy import meshgrid
from numpy.random import seed
from numpy.random import rand
from matplotlib import pyplot

# objective function
def objective(x, y):
    return x**2.0 + y**2.0

# define range for input
r_min, r_max = -5.0, 5.0

# sample input range uniformly at 0.1 increments
xaxis = arange(r_min, r_max, 0.1)
yaxis = arange(r_min, r_max, 0.1)

# create a mesh from the axis
x, y = meshgrid(xaxis, yaxis)

# compute targets
results = objective(x, y)

# simulate a sample made by an optimization algorithm
seed(1)
sample_x = r_min + rand(10) * (r_max - r_min)
sample_y = r_min + rand(10) * (r_max - r_min)

# create a filled contour plot with 50 levels and jet color scheme
pyplot.contourf(x, y, results, levels=50, cmap='jet')

# define the known function optima
optima_x = [0.0, 0.0]

# draw the function optima as a white star
pyplot.plot([optima_x[0]], [optima_x[1]], '*', color='white')

# plot the sample as black circles
pyplot.plot(sample_x, sample_y, 'o', color='black')

# show the plot
pyplot.show()
```
