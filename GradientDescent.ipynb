```python
from numpy.random import rand

def gradient_descent(objective, derivative, bounds, n_iter, step_size):
    # generate an initial point
    solution = bounds[:, 0] + rand(len(bounds)) * (bounds[:, 1] - bounds[:, 0])
    # run the gradient descent
    for i in range(n_iter):
        # calculate gradient
        gradient = derivative(solution)
        # take a step
        solution = solution - step_size * gradient
        # evaluate candidate point
        solution_eval = objective(solution)
        # report progress
        print('>%d f(%s) = %.5f' % (i, solution, solution_eval))
    return [solution, solution_eval]
```

```python
from numpy import asarray, arange
from numpy.random import rand
from matplotlib import pyplot as plt

# Función objetivo
def objective(x):
    return x**2.0

# Derivada de la función objetivo
def derivative(x):
    return 2.0 * x

# Algoritmo de descenso por gradiente
def gradient_descent(objective, derivative, bounds, n_iter, step_size):
    # listas para guardar soluciones y evaluaciones
    solutions, scores = list(), list()
    # punto inicial aleatorio en el rango
    solution = bounds[:, 0] + rand(len(bounds)) * (bounds[:, 1] - bounds[:, 0])
    # ciclo de iteraciones
    for i in range(n_iter):
        gradient = derivative(solution)
        solution = solution - step_size * gradient
        solution_eval = objective(solution)
        solutions.append(solution.item())  # convertir arreglo a escalar
        scores.append(solution_eval.item())
        print(f'>{i} f({solution}) = {solution_eval.item():.5f}')
    return solutions, scores

# Definir rango de entrada
bounds = asarray([[-1.0, 1.0]])
n_iter = 30
step_size = 0.1

# Ejecutar descenso por gradiente
solutions, scores = gradient_descent(objective, derivative, bounds, n_iter, step_size)

# Generar puntos para graficar la función objetivo
inputs = arange(bounds[0, 0], bounds[0, 1] + 0.1, 0.1)
results = objective(inputs)

# Graficar función objetivo (línea azul)
plt.plot(inputs, results, label='Objective function')

# Graficar puntos de descenso por gradiente (puntos rojos conectados)
plt.plot(solutions, scores, '.-', color='red', label='Gradient descent')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent on $x^2$')
plt.legend()
plt.show()
```
